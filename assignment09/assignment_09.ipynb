## define the feature functions

- feature vector is defined by $(1, f_1(x, y), f_2(x, y), \cdots, f_{k-1}(x, y)) \in \mathbb{R}^k$

```python
def compute_feature1(point):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    one = np.ones(shape=(len(point[:,0]),))
    x   = point[:,0]
    xx  = np.power(x,2)
    y   = point[:,1]
    yy  = np.power(y,2)
    
    feature = np.c_[one,x,y,xx,yy]
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return feature
```

```python
def compute_feature2(point):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    one = np.ones(shape=(len(point[:,0]),))
    x   = point[:,0]
    xx  = np.power(x,2)
    xxx  = np.power(x,3)
    y   = point[:,1]
    yy  = np.power(y,2)
    yyy  = np.power(y,3)
    
    feature = np.c_[one,x,y,xx,yy,xxx,yyy]
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return feature
```

## define the linear regression function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature = $(1, f_1(x, y), \cdots, f_{k-1}(x, y)) \in \mathbb{R}^k$

```python
def compute_linear_regression(theta, feature):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    value = np.dot(feature,theta.T)
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return value
```

## define sigmoid function with input

- $z \in \mathbb{R}$

```python
def sigmoid(z):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    value = 1 / (1 + np.exp(-z))
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return value 
```

## define the logistic regression function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$

```python
def compute_logistic_regression(theta, feature):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    value = sigmoid(compute_linear_regression(theta, feature))
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return value
```

## define the residual function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_residual(theta, feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    h = compute_logistic_regression(theta, feature)
    residual = -((label)*np.log(h) + (1-label)*np.log(1-h))
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return residual
```

## define the loss function for the logistic regression

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_loss(theta, feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    loss = 1 / len(label) * np.sum(compute_residual(theta, feature, label))
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return loss
```

## define the gradient of the loss with respect to the model parameter $\theta$

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_gradient(theta, feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    gradient = 1 / len(label) * np.dot(feature.T, (compute_logistic_regression(theta,feature) - label))
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return gradient
```

## compute the accuracy of the prediction for point with a given model parameter

```python
def compute_accuracy(theta, feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    accuracy = 0
    predict_label = compute_logistic_regression(theta, feature)
    for i in range(len(predict_label)):
        if (predict_label[i] >= 0.5 and label[i] == 1) or (predict_label[i] < 0.5 and label[i] == 0):
            accuracy += 1
        
    accuracy = accuracy / len(label)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return accuracy
```

## initialize the gradient descent algorithm

```python
data1_number_iteration   = 10000
data2_number_iteration   = 10000

data1_learning_rate   = 0.1 
data2_learning_rate   = 0.1 

data1_number_feature    = 5
data2_number_feature    = 7

theta1  = np.zeros(data1_number_feature)
theta2  = np.zeros(data2_number_feature)

data1_loss_iteration  = np.zeros(data1_number_iteration)
data2_loss_iteration  = np.zeros(data2_number_iteration)

data1_accuracy_iteration    = np.zeros(data1_number_iteration)
data2_accuracy_iteration    = np.zeros(data2_number_iteration)
```

## run the gradient descent algorithm to optimize the loss function with respect to the model parameter

```python
for i in range(data1_number_iteration):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #

    feature1    = compute_feature1(data1_point)
    label1      = data1_label
    
    theta1      = theta1 - data1_learning_rate * compute_gradient(theta1,feature1,label1)
    loss1       = compute_loss(theta1,feature1,label1)
    accuracy1   = compute_accuracy(theta1, feature1, label1)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    data1_loss_iteration[i]     = loss1
    data1_accuracy_iteration[i] = accuracy1

data1_theta_optimal = theta1
```

```python
for i in range(data2_number_iteration):
    
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    feature2    = compute_feature2(data2_point)
    label2      = data2_label
    
    theta2      = theta2 - data2_learning_rate * compute_gradient(theta2,feature2,label2)
    loss2       = compute_loss(theta2,feature2,label2)
    accuracy2   = compute_accuracy(theta2, feature2, label2)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    data2_loss_iteration[i]     = loss2
    data2_accuracy_iteration[i] = accuracy2

data2_theta_optimal = theta2
```
